```python
import pandas
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from statannot import add_stat_annotation
from tqdm import tqdm
import copy
```

# Load data

```python
data = pandas.read_csv(r'Data/Fig_1_Diameter_Circularity.csv')
```

# Area and circularity at 24h and 48h

```python
# Define font
plt.rc('font',family='Arial')

# Plot
fig, ax = plt.subplots(1, figsize = (1.8,3))

ax.tick_params(axis="x", labelsize=14) 
ax.tick_params(axis="y", labelsize=14) 

order = [24, 48]

ax.set_ylim(100,350)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.tick_params(axis='x', labelsize=14, rotation=45)
ax.tick_params(axis='y', labelsize=14)        
ax.tick_params(width=1, length=4)
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(1)
           
sns.violinplot(
    x = 'time',
    y = 'D',
    data=data,
    order = order,
    palette = ['silver', 'silver'],
    errorbar='se',
    errcolor='black',
    capsize=0.2,
    errwidth=1,
    color = 'r',
    lw = 1, 
    edgecolor="0",
    inner = 'quartiles',
)

add_stat_annotation(ax,
                    data=data,
                    x='time',
                    y='D',
                    order = order,
                    fontsize = 16,
                    box_pairs=[(24,48)],
                    line_offset_to_box = 0.15,
                    test='t-test_welch', comparisons_correction=None, 
                    text_format='star', loc='inside', verbose=0)

ax.set_ylabel(r'Diameter (Âµm)', fontsize=14)
ax.set_xlabel('Time (h)', fontsize=14)
plt.tight_layout()
```
