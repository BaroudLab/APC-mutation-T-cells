```python
import pandas
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from statannot import add_stat_annotation
import copy
```

# Load data

```python
data_brdu = pandas.read_csv(r'Data\Fig_S1_BrdU.csv')
data_caspase = pandas.read_csv(r'Data\Fig_S1_Caspase.csv')
data_pi = pandas.read_csv(r'Data\Fig_S1_PI.csv')
```

# Brdu chip 24h

```python
# Define font
plt.rc('font',family='Arial')

# Plot
fig, ax = plt.subplots(1, figsize = (1.5,3))

order = [24]
ax.set_ylim(0,25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.tick_params(axis='x', labelsize=14, rotation=45)
ax.tick_params(axis='y', labelsize=14)        
ax.tick_params(width=1, length=4)
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(1)
            
sns.barplot(
    x = 'time',
    y = 'percent_div',
    data=data_brdu,
    order = order,
    palette = ['lime'],
    errorbar='se',
    errcolor='black',
    capsize=0.2,
    errwidth=1,
    color = 'r',
    width = 0.5,
    lw = 1, 
    edgecolor="0"
)

ax.set_ylabel('Brdu+ (%)', fontsize = 14)
ax.set_xlabel('Time (h)', fontsize = 14)
plt.tight_layout()
```
